<h2>logistic function</h2>

```python
from matplotlib import pyplot as plt
import numpy as np 
import math as m
```

```python
x = np.arange(-10,10,dtype = float)
y = 1/(1+np.power(m.e,-x))
```

```python
#折线图使用plot绘制
plt.title("logistic function")
plt.plot(x,y)
plt.xticks(np.arange(-15,16,2))#将列表绘制在坐标轴上
plt.yticks(np.arange(0,11)/10)
plt.show()
```

<h2>温度变化趋势</h2>

```python
x = range(0,121)
y = [random.randint(20,35) for i in range(121)]
print(y)
plt.plot(x,y)
_xticks_labels = ["10点{}分".format(i) for i in range(60)]
_xticks_labels += ["11点{}分".format(i) for i in range(60)]
_xticks_labels += ["12点00分"]
plt.xticks(x[::5],_xticks_labels[::5],rotation=-90)#数字与字符串对应,步长为5
#rotation是旋转的度数
plt.yticks(np.arange(20,np.amax(y)+2))
plt.xlabel("时间")
plt.ylabel("温度 单位(摄氏度")
plt.title("10点到12点每隔1分钟的温度变化情况")
plt.figure(figsize=(15,15),dpi=90)#调整画幅与清晰度
plt.show()
```

<h2>绘制多条折线图</h2>

```python
a=[random.randint(0,6) for i in range(20)]
b=[random.randint(0,3) for i in range(20)]
xa = [i for i in range(11,31)]
xb = range(11,31)
_xa=["{}岁".format(i) for i in xa]
_xb=["{}岁".format(i) for i in xb]
plt.figure(figsize=(10,8),dpi=80)#必须在绘画前面写
plt.plot(xa,a,label="自己", color="orange",linestyle='--')#给线注名,并添加颜色
plt.plot(xa,b,label="同桌", color="cyan")#在同一画布下绘制两条线
#同样可以设置线条风格 linestyle = '--'线条风格
#linewidth=5线条粗细
#alpha=0.5线条透明度
plt.xticks(xa,_xa,rotation=-90)
plt.xlabel("年龄")
plt.ylabel("个数")
plt.title("趋势图")

#绘制网格
plt.grid(alpha=0.4)#alpha,调整网格的透明度
plt.yticks(range(0,8))
plt.legend(loc="upper left")
plt.show()
```
